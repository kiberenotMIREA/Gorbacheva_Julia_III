Импортируем библиотеку NumPy

```python
import numpy as np
```

Создаем массив из списка

```python
arr_from_list = np.array([1, 2, 3, 4, 5])
print("Массив из списка:", arr_from_list)
```

Создаем двумерный массив из вложенного списка

```python
matrix_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("Двумерный массив:\n", matrix_2d)
```

Создаем массив из нулей

```python
zeros_arr = np.zeros(5)
print("Массив из нулей:", zeros_arr)
```

Создаем двумерный массив из единиц

```python
ones_matrix = np.ones((2, 3))
print("Массив из единиц (2x3):\n", ones_matrix)
```

Создаем массив с последовательностью чисел от 0 до 10 с шагом 2

```python
range_arr = np.arange(0, 10, 2)
print("Массив с arange:", range_arr)
```

Выводим размерность массива

```python
print("Размерность matrix_2d:", matrix_2d.shape)
```

Выводим тип элементов массива

```python
print("Тип данных matrix_2d:", matrix_2d.dtype)
```

Выводим исходный одномерный массив

```python
print("Исходный массив:", arr_from_list)
```

Выводим первый элемент

```python
print("Первый элемент:", arr_from_list[0])
```

Выводим элементы с индексом 1 до 3

```python
print("Срез [1:4]:", arr_from_list[1:4])
```

Выводим каждый второй элемент

```python
print("Срез с шагом 2:", arr_from_list[::2])
```

Выводим исходный двумерный массив

```python
print("\nДвумерный массив:\n", matrix_2d)
```

Выводим 1 строку, 2 столбец

```python
print("Элемент [1, 2]:", matrix_2d[1, 2])
```

Выводим все строки, столбец 1

```python
print("Столбец 1:", matrix_2d[:, 1])
```

Выводим подматрицу 2x2

```python
print("Срез строк:\n", matrix_2d[0:2, 0:2])
```

Создаем массивы a и b

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

Выводим поэлементное сложение

```python
print("Сложение a + b:", a + b) 
```

Выводим поэлементное умножение

```python
print("Умножение a * b:", a * b)
```

Выводим среднее значение

```python
print("Среднее значение:", np.mean(a))
```

Выводим стандартное отклонение

```python
print("Стандартное отклонение:", np.std(a))
```

Выводим максимальный элемент

```python
print("Максимальный элемент:", np.max(a))
```

Выводим исходный двумерный массив

```python
print("Двумерный массив:\n", matrix_2d)
```

Выводим среднее по всему двумерному массиву

```python
print("Среднее по всему массиву:", np.mean(matrix_2d))
```

Выводим среднее по столбцам

```python
print("Среднее по столбцам:", np.mean(matrix_2d, axis=0))
```
